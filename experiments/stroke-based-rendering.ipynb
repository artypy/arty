```python
from arty.sbr.color import visualize_color_set
from arty.sbr.preset import Preset
from arty.sbr.image_painter import ImagePainter
```

```python
import cv2

preset = Preset(stroke_scale=1,
                gradient_smoothing_radius=20,
                length_scale=1/3,
                brush_type="ellipse",
                gradient_type="sharr",
                gradient_smoothing_type="gaussian",
                smoothing_iterations=1,
                length="base",
                length_first_flag=True,
                grid_scale=2,
                grayscale=False,
                has_cardboard=True,
                palette_size=10)

import yaml

with open(preset.preset_save_path, 'w') as file:
    yaml.dump(preset.__dict__, file)

with open(preset.preset_save_path) as file:
    preset = Preset(**{k: v for k, v in yaml.load(file, Loader=yaml.FullLoader).items() if
                       k not in ["img_save_path", "preset_save_path"]})
```

```python
img = cv2.imread(preset.img_path)

painter = ImagePainter(img, preset)
painter.prepare_color_set()
painter.compute_gradient()
result = painter.paint()
cv2.imwrite(preset.img_save_path, result)
painter.show_result()
```

```python
visualize_color_set(painter.color_set)
```
