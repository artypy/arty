```python
from arty import filter
import core
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
img = (plt.imread('images/img.png') * 255).astype(int)
plt.imshow(img)
plt.show()
```

```python
# dx, dy = core.edge.gradientX(img), core.edge.gradientY(img)
import cv2
from core.edge.gradient import Gradient
import numpy as np

image = img.astype(np.uint8)
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gradient = Gradient(image)
dx, dy = gradient.grad_x, gradient.grad_y

edges = core.edge.gradient(img)
angles = core.angle.gradient_truth(dx, dy)
lines = core.filter.lines(img, angles, edges)

# lines = filter.lines(img)

plt.imshow(lines)
plt.show()
```
