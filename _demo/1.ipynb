```python
from arty import core
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10, 10)
```

```python
img = (plt.imread('images/img.png') * 255).astype(int)
plt.imshow(img)
plt.show()
```

```python
dx, dy = core.edge.gradientX(img), core.edge.gradientY(img)
edges = core.edge.gradient(img)
angles = core.angle.gradient_truth(dx, dy)
lines = core.filter.lines(img, angles, edges)

plt.imshow(lines)
plt.show()
```

```python
blur = core.filter.blur.gauss(img)
dx, dy = core.edge.gradientX(blur), core.edge.gradientY(blur)
edges = core.edge.gradient(blur)
angles = core.angle.gradient_truth(dx, dy)
lines_blur = core.filter.lines(blur, angles, edges)

plt.imshow(lines_blur)
plt.show()
```

```python
noised = core.filter.noise(blur)
lines_noise = core.filter.lines(noised, angles, edges)

plt.imshow(lines_noise)
plt.show()
```
